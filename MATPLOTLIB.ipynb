# Pengenalan Matplotlib

* Matplotlib adalah salah satu module dalam pemrograman Python yang digunakan untuk visualisasi data dalam bentuk graph.
* Graph menggunakan Matplotlib digambarkan pada suuatu `figure`, yang di dalamnya terdiri dari satu atau lebih `axes`.

## Import Library

```python
%matplotlib inline
```

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Plotting Sederhana

Kita dapat menyimpan hasil visualisasi kita dalam suatu `figure` yang memiliki sebuah `axes`, kemudian melakukan plotting data pada `axes`.

```python
# Data sampel
x = np.array([1, 3, 5, 7 ])
y = np.array([10, 1, 3, 4])

# Membuat sebuah figure dan sebuah axes
fig, ax=plt.subplots()

# Melakukan plotting data pada axes
ax.plot(x,y)
```

Cara lain untuk melakukan plotting adalah dengan menggunakan method `plot` pada `pyplot` untuk melakukan plotting.

```python
# Data sampel
x = np.array([1, 3, 5, 7 ])
y = np.array([2, 1, 3, 4])

# Melakukan plotting 
plt.plot(x,y) 
```

## Komponen dari sebuah Matplotlib `Figure`

* `Figure` adalah media dasar untuk melakukan visualisasi menggunakan Matplotlib
* Dalam sebuah `figure` kita dapat memiliki beberapa `axes` 
* `axes` adalah area gambar dalam sebuah `figure`
* `axes` berada di dalam `figure`
* `axis` sumbu dalam dimensi visualisasi data. Jumlah `axis` bisa dua sampai tiga yaitu `x, y ,z`

## Mengenal Object Oriented Style untuk melakukan ploting

Metode ini secara eksplisit membuat `figures` beserta `axes`, kemudian memanggil method dari keduanya. `figures` dan `axis` dianggap sebuah object dalam bentuk vairabel.

```python
# Membuat sample data
x = np.linspace(0, 2, 100)
x
```

```python
# Membuat sebuah figure dan sebuah axes
fig, ax = plt.subplots()

# Plotting tiga variabel data pada axes
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='kuadrat')
ax.plot(x, x**3, label='kubik')

# Menambahkan label sumbu x pada axes
ax.set_xlabel('label x')

# Menambahkan label sumbu y pada axes
ax.set_ylabel('label y')

# Menambahkan judul pada axes
ax.set_title('Plot sederhana')

# Menambahkan legend pada axes
ax.legend()
```

## Membuat plotting sederhana

```python
plt.plot([2, 4, 1, 7])
plt.ylabel('sumbu y')
plt.show()
```

* Pemanggilan fungsi `plot()` dapat dilakukan dengan menyertakan hanya sebuah deret bilangan (`list/array`) ataupun dua buah deret bilangan (`list/array`)
* Bia pemanggilannya hanya menyertakan sebuah deret bilangan, maka nilai pada deret bilangan tersebut akan dijadikan data point untuk nilai pada sumbu y, sedangkan nilai pada sumbu x akan secara otomatis dibuatkan sebuah deret bilangan terurut dengan nilai dimulai dari nol(indeks).

```python
plt.plot([1, 3, 5, 7],[2, 8, 5, 7])
plt.show()
```

## Pengaturan Format pada plot

* Selain menyertakan dua buah deret bilangan sebagai parameter untuk sumbu x dan y terdapat parameter ketiga yang bida kita sertakan untuk mengatur warna dan jenis plotting.
* Pengaturan parameter ketiga ini sepenuhnya mengadopsi gaya formating pada MATLAB
* Nilai defaults parameter ketiga ini adalah `b__`.
* Pemanggilan fungsi `axis()` menyertakan sebuah `list[xmin, xmax, ymin, ymax]`

```python
plt.plot([1,2,3,4],[1,4,8,10],'r*')
plt.show()
```

## Multiplot dalam satu pemanggilan fungsi `plot()`

```python
t=np.arange(0., 5., 0.2)
t
```

```python
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
```

Note: Butuh catatan tentang macam-macam kode untuk bentuk dan warna

Kita juga dapat melakukan multiplot dalam satu pemanggilan fungsi `plot()`

```python
plt.plot(t, t, 'r--',
        t, t**2, 'bs',
        t, t**3, 'g^')
plt.show()
```

## Plotting dengan keywords

Matplotlib juga memungkinkan kita untuk melakukan plotting sekumpulan nilai yang tersimpan dalam struktur data yang diserta keywords argument(kwargs) seperti pada `dictionary` dan `Pandas DataFrame`

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}

data['b']=data['a']+10*np.random.randn(50)
data['d']=np.abs(data['d'])*100

data
```

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
plt.scatter('a', 'b',c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

# Multiple Subplots

Dalam sesi ini kita akan mempelajari:
* Simple Line Plot
* Multiple Subplots dengan OO Stye
* Multiple Subplots dengan pyplot style

## Simple Line Plot

```python
x=np.arange(0.0, 2.0, 0.01)
x
```

```python
s = np.sin(2*np.pi*x)
s
```

```python
fig, ax=plt.subplots()

ax.plot(x,s)
ax.set(xlabel='nilai x',
      ylabel='nilai sin(x)',
      title='Visualisasi Fungsi Sin(x)')
ax.grid()

plt.show()
```

## Multiple Subplots

```python
x1=np.linspace(0.0, 5.0, 100)
x2=np.linspace(0.0, 2.0, 100)

y1=np.cos(2*np.pi*x1)
y2=np.cos(2*np.pi*x2)
```

```python
x1
```

## Multiple Subplots dengan OO Style

```python
fig, (ax1, ax2)=plt.subplots(2,1)
fig.suptitle("Multiple Subplots")

ax1.plot(x1,y1, 'ro-')
ax1.set_ylabel('Nilai cos(x1)')

ax2.plot(x2,y2, 'g.-')
ax2.set_ylabel('Nilai cos(x2)')

ax2.set_xlabel('Nilai x')

plt.show()
```

# Histogram 

Dalam sesi ini kita akan membuat histogram dengan Matplotlib

## Sample Data

```python
# Nilai Rataa-rata dan Standar Deviasi
mu, sigma = 100, 15

# Distribusi Normal
x = mu + sigma * np.random.randn(10000)

x
```

```python
x.shape
```

## Histogram dengan Pyplot Style

```python
plt.hist(x,
        bins=100,
        facecolor='g',
        alpha=0.75)
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45, 500, r'$\mu=100,\ \sigma=15$')
plt.grid()

plt.show()
```

## Histogram dengan OO Style

```python
fig, ax=plt.subplots()

ax.hist(x,
        bins=50,
        facecolor='g',
        alpha=0.75)
ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram')

ax.text(45, 500, r'$\mu=100,\ \sigma=15$')
ax.grid()

plt.show()
```

# Bar Plot / Bar Chart

## Simple Bar Plot

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumah')
plt.title('Contoh Bar Plot')

plt.show()
```

## Pengaturan Grid dan Color

```python
plt.bar(kategori, 
        data,
       color='red',
       alpha=0.25)

plt.grid(linestyle='--',
         linewidth=2,
         axis='y', #pilihan = x, y, both
         alpha=0.5)

plt.xlabel('Kategori')
plt.ylabel('Jumah')
plt.title('Contoh Bar Plot')

plt.show()
```

## Stacked Bar Plot

```python
data1=[25, 85, 75, 40, 60]
data2=[40, 35, 20, 55, 10]
kategori=['A', 'B', 'C', 'D', 'E']
```

```python
# Plot Data Pertama
plt.bar(kategori, data1, label='Data1')
#plt.bar(kategori, data1, data2, label='Data2')
# Plot Data Kedua
plt.bar(kategori, data2, label='Data 2', bottom=data1)

plt.legend()

plt.grid(linestyle='--',
         linewidth=2,
         axis='y', #pilihan = x, y, both
         alpha=0.5)

plt.xlabel('Kategori')
plt.ylabel('Jumah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

## Grouped Bar Plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']

x=np.arange(len(kategori))
x
```

```python
width= 0.35

plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori)

plt.legend()

plt.grid(linestyle='--',
         linewidth=1,
         axis='y', #pilihan = x, y, both
         alpha=0.5)

plt.xlabel('Kategori')
plt.ylabel('Jumah')
plt.title('Contoh Grouped Bar Plot')

plt.show()
```

## Matplotlib Style

### Fungsi Plotting

```python
def gambar_barplot():
    data1=[25, 85, 75, 40, 60]
    data2=[40, 35, 20, 55, 10]
    kategori = ['A', 'B', 'C', 'D', 'E']
    
    x=np.arange(len(kategori))
    
    width= 0.35
    
    plt.bar(x-width/2, data1, width, label='Data 1')
    plt.bar(x+width/2, data2, width, label='Data 2')
    
    plt.xticks(x, kategori)
    
    plt.grid(linestyle='--',
         linewidth=1,
         axis='y', #pilihan = x, y, both
         alpha=0.5)
    
    plt.xlabel('Kategori')
    plt.ylabel('Jumah')
    plt.title('Contoh Grouped Bar Plot')
    
    plt.legend()
    plt.show()
```

```python
gambar_barplot()
```

## Box Plot/ Box Chart

### Sample Dataset

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

## Simple Box Plot

```python
plt.boxplot(data, labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

### Notched Box Plot

```python
plt.boxplot(data, labels=['Data'], notch=True)

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

## Pengaturan Marker Untuk Outlier

```python
plt.boxplot(data, 
            labels=['Data'], 
            showfliers=True,
            flierprops={'markerfacecolor':'g', 'marker':'^'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

## Horizontal Boxplot

```python
plt.boxplot(data, 
            labels=['Data'], 
            vert=False)

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

## Twin Axes

```python
x = np.linspace(1, 10, 25)
x
```

```python
fig, ax1=plt.subplots()

ax1.plot(x, np.exp(x), 'bs-', label='exp')
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('Sumbu Y')

ax2=ax1.twinx()
ax2.plot(x, np.log(x), 'ro-', label='log')
ax2.set_ylabel('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc='upper left')

plt.show()
```
